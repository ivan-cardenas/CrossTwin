```python
from owslib.wfs import WebFeatureService
```

```python
getCapabilities = 'service=WFS&request=GetCapabilities'
url = 'https://service.pdok.nl/cbs/wijkenbuurten/2024/wfs/v1_0?'+getCapabilities
```

```python
wfs = WebFeatureService(url)
```

```python
for name, layer in wfs.contents.items():
    print(f"Name: {name}, Layer: {layer}")
    print(f"Title: {layer.title}")
    print("-"*50)
```

```python
response = wfs.get_schema('cbs:Wijken')
print(response)
```
